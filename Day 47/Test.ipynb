```python
from sklearn import datasets
from sklearn import linear_model
import sklearn as skl
import matplotlib.pyplot as plt
```

```python
diabetes = datasets.load_diabetes()
data = diabetes.data
target = diabetes.target
```

```python
regr = linear_model.LinearRegression()
# We call it "regr" because this time it's
# a regressor instead of a classifier
```

```python
data[400]
# 10 predictive values
```

```python
target[400]
# Measure of disease progression
```

```python
split = 425
regr.fit(data[:split], target[:split])
```

```python
make_graph = lambda: plt.plot(
    regr.predict(data[split:]),
    linestyle=':',
) + plt.plot(
    target[split:],
)
make_graph()
# Blue is the prediction, orange is the reality
```

```python
# Okay I'm not the best at making graphs, but it's good enough,
# as we can see, the predictions aren't perfect, but they seem
# a lot better than chance
```
